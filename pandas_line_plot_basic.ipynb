# การสร้างกราฟเส้นด้วย pandas เบื้องต้น

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter # for custom number format on x and y axis
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
print(f'pandas version: {pd.__version__}')
```

```python
pd.set_option('display.max_rows', 10)
```

```python
oil=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/OPEC_monthly_oilprice.csv',
                index_col='Date', parse_dates=True)
oil
```

```python
oil.plot(kind='line')
```

```python
oil.plot(kind='line');
```

```python
ax = oil.plot(kind='line', title='OPEC oil price US$/barrel')
ax.set_xlabel('Date')
ax.set_ylabel('US$/barrel');
```

https://matplotlib.org/2.0.0/examples/color/named_colors.html

```python
ax = oil.plot(kind='line', title='OPEC oil price US$/barrel', color='chocolate')
ax.set_xlabel('Date')
ax.set_ylabel('US$/barrel');
```

```python
oil['Oil'].mean()
```

```python
ax = oil.plot(kind='line', title='OPEC oil price US$/barrel', color='maroon')
ax.set_xlabel('')
ax.set_ylabel('US$/barrel')
ax.axhline(oil['Oil'].mean(), color="yellowgreen", linestyle="--", linewidth=.8);
```

```python
ax = oil.plot(kind='line', title='OPEC oil price US$/barrel', color='maroon', figsize=(10, 5))
ax.set_xlabel('')
ax.set_ylabel('US$/barrel')
ax.axhline(oil['Oil'].mean(), color="yellowgreen", linestyle="--", linewidth=.8);
```

```python
ax = oil.loc['2015-01-31':, :].plot(kind='line', title='OPEC oil price US$/barrel', 
                                    color='maroon', figsize=(10, 5))
ax.set_xlabel('')
ax.set_ylabel('US$/barrel')
ax.axhline(oil.loc['2015-01-31':, 'Oil'].mean(), 
           color="yellowgreen", 
           linestyle="--", linewidth=.8);
```
